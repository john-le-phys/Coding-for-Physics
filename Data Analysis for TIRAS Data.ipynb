```python
def phase_shift_angular(a,b,r):
    c = np.sqrt(a**2+b**2+(2*a*b*r))
    return c

#let a be the unblocked voltages and b be the blocked voltages
a, b = (xdata,ydata) = loadtxt('a and b data.txt', unpack=True)

signal = phase_shift_angular(a,b,r)
r = radians(degrees)
degrees = xdata
a, b = (xdata,ydata) = loadtxt('a and b data.txt', unpack=True)

print(signal)
```

```python

# importing numpy and matplotlib
import numpy as np
import matplotlib.pyplot as plt
 
(x_1data) = loadtxt('kilohertz.txt', unpack=True)
(x_2data,y_2data)=(x_1data,signal)
print(x_2data,y_2data)
 
# creating graph space for two graphs
graph, (plot2) = plt.subplots(1, 1)

 
# plot2 graph for inverted axes
plot2.plot(x_2data, y_2data)
plot2.set_title("Concentraion of Solvated Electrons Per Kilohertz")
plot2.invert_xaxis()
plt.xticks(x_2data)
plot2.set_xlabel('Kilohertz')
plot2.set_ylabel('Concentration of Solvated Electrons-Millivlots')
 
# display the graph
graph.tight_layout()
plt.show()
```
